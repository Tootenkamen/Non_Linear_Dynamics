```python
import numpy as np
import matplotlib.pyplot as plt
from math import log

class Henon:
    
    #Defining the equation for Henon Map
    def HM(self,p,a,b):
        return np.array([1 - a * p[0]**2 + p[1] , b * p[0]])
    
    #Generating the points using the formula
    def Gen(self,p,a,b,n):
        self.p = p
        self.a = a
        self.b = b
        self.n = n
        self.pcoll = [p]
        for self.i in range(n):
            self.p = self.HM(self.p,a,b)
            self.pcoll.append(self.p)
        self.pcoll = np.array(self.pcoll)
        return self.pcoll
    
    #Plotting the points
    def Plot(self):
        plt.scatter(self.pcoll[:,0],self.pcoll[:,1])
        plt.title(f'Points of a Henon Map with Iterations {self.n} , a={a},b={b}')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
    
class GP:
    
    #Heaviside Step Function
    def HS(self,x):
        if x<0:
            return 0
        elif x==0:
            return 0.5
        else:
            return 1
    
    #Finding the distance between the given points
    def N(self,xi,xj):
        return np.linalg.norm(xi-xj)
    
    #Applying the Heaviside step function to the difference between the given r and distnace between two points
    def Theta(self,r,xi,xj):
        return self.HS(r-self.N(xi,xj))
    
    #Calculating the Correlation Dimension
    def Dim(self, r, p): 
        self.sum = 0 
        self.n = p.shape[0]
        for self.i in range(self.n):
            for self.j in range(self.i):
                self.sum += self.Theta(r,p[self.i],p[self.j])
        return (self.sum * 2) / ((self.n)*(self.n-1)) 
    
    
def Plot_C(r,fit):
    r = np.arange(0.15,6,0.05) 
    c = []
    d = []
    for i in r:
        CD = Alg.Dim(i,Pts)
        c.append([i,CD])
        d.append([np.log10(i),np.log10(CD)])
    c = np.array(c)
    d = np.array(d)

    #Plotting  C vs r 
    plt.plot(c[:,0],c[:,1])
    plt.xlabel('r')
    plt.ylabel('C(r)')
    plt.title('C vs r')
    plt.show()

    #Plotting  log(C) vs log(r) 
    plt.plot(d[:,0],d[:,1])
    plt.xlabel('log(r)')
    plt.ylabel('log(C)')
    plt.title('log(C) vs log(r)')
    plt.show()

    if fit==True:
        #Plotting  log(C) vs log(r) 
        plt.scatter(d[:,0],d[:,1])
        m , c = np.polyfit(d[:-100,0],d[:-100,1],1)
        y = m*d[:,0] + c
        plt.plot(d[:,0],y)
        plt.xlabel('log(r)')
        plt.ylabel('log(C)')
        plt.title(f'Dimension = dy/dx ={m}')
        plt.show()


  









```
